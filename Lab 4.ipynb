<h2>Завдання № 5 Ввести 2 положительных числа n и m (n < m). Вывести на экран сумму чисел между n и m, то есть не включая числа n и m в сумму.</h2>

```python
n = int(input("Введите число n = "))
m = int(input("Введите число m = "))
counter = 0
for i in range (n + 1, m):
    counter += i
print("Ответ = ", counter)
```

<h2>Завдання № 7 Вычислить значение выражения y = sin(x) + sin^2(x) + sin^3(x) + sin^n(x), где n и х задаются с клавиатуры.
</h2>

```python
import math
n = int(input("Введите n = "))
x = int(input("Введите x = "))
for i in range(1, n):
    print(i)
y = math.sin(x) + math.sin(x)**2 + math.sin(x)**3 + math.sin(x)**n
print("Ответ = ",y)

        
```

<h2>Завдання № 9 Написать функцию для подсчета каждого значения от 1 до num. num Всегда положительное целое число.</h2>

```python
while True:
  output = ""
  num = int(input("Введите целое число: "))

  if num == 0:
    exit()

  for i in range(1, num+1):
    output += "{}".format(i)
    if i != num:
      output += "+"
  output += " = {}".format(sum(range(num+1)))
  print (output)
```
